## Data Attributes

Age: Numerical value representing the age of the individual.

Gender: Categorical value indicating the gender of the individual (Male/Female).

Tumor Size: Numerical value indicating the size of the tumor in centimeters.

Tumor Grade: Categorical value indicating the grade of the tumor (High, Medium, Low, NaN for missing data).

Symptoms Severity: Categorical value representing the severity of symptoms (Mild, Moderate).

Family History: Binary value indicating whether there is a family history of cancer (Yes/No).

Smoking History: Categorical value indicating the smoking history of the individual (Current Smoker, Former Smoker,Non-Smoker,nan).

Alcohol Consumption: Categorical value indicating the level of alcohol consumption (High, Moderate, low, NaN for missing data).

Exercise Frequency: Categorical value representing the frequency of exercise (Regularly, Rarely, Occasionally,Never, NaN for missing data).

Cancer Present: Binary value indicating whether cancer is present (1) or not (0).

```python
## types of columns in the data

numeric=['Age','Tumor_Size']
ordinal=['Tumor_Grade','Symptoms_Severity','Alcohol_Consumption','Exercise_Frequency']
nominal=['Smoking_History','Gender','Family_History']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,OrdinalEncoder,OneHotEncoder
```

```python
numeric_preprocess=Pipeline([('Mean Imputation',SimpleImputer(strategy='mean')),('Scaling',StandardScaler())])
numeric_preprocess
```

```python
ordinal_preprocess=Pipeline([('Mode Imputation',SimpleImputer(strategy='most_frequent')),('Encoding',OrdinalEncoder())])
ordinal_preprocess
```

```python
nominal_preprocess=Pipeline([('Model Imputation',SimpleImputer(strategy='most_frequent')),('Encoding',OneHotEncoder())])
nominal_preprocess                      
```

### Preprocessing

```python
preprocess=ColumnTransformer([('numerical transformation',numeric_preprocess,numeric),('ordinal transformation',ordinal_preprocess,ordinal),('nominal transformation',nominal_preprocess,nominal)])
preprocess
```

### 3. Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
X=df.drop(columns='Cancer_Present',axis=1)
```

```python
y=df['Cancer_Present']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=20)
```

#### Decision tree

```python
dt=Pipeline([('Data preprocessing',preprocess),('Algorithm',DecisionTreeClassifier())])
dt.fit(X_train,y_train).score(X_test,y_test)
```

#### Logistic Regression

```python
log=Pipeline([('Data Preprocessing',preprocess),('Algorithm',LogisticRegression())])
log.fit(X_train,y_train).score(X_test,y_test)
```

#### KNN

```python
kn=Pipeline([('Data Preprocessing',preprocess),('Algorithm',KNeighborsClassifier())])
kn.fit(X_train,y_train).score(X_test,y_test)
```

#### SVM

```python
svm=Pipeline([('Data Preprocessing',preprocess),('Algorithm',SVC())])
svm.fit(X_train,y_train).score(X_test,y_test)
```

#### XG boost

```python
from xgboost import XGBClassifier
```

```python
xg=Pipeline([('Data Preprocessing',preprocess),('Algorithm',XGBClassifier())])
xg.fit(X_train,y_train).score(X_test,y_test)
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf=Pipeline([('Data Preprocessing',preprocess),('Algorithm',RandomForestClassifier())]) 
rf.fit(X_train,y_train).score(X_test,y_test)
```

#### Model Deployment

```python
import pickle
pickle.dump(dt,open('dt.pkl','wb'))
pickle.dump(log,open('log.pkl','wb'))
pickle.dump(kn,open('kn.pkl','wb'))
pickle.dump(svm,open('svm.pkl','wb'))
pickle.dump(xg,open('xg.pkl','wb'))
pickle.dump(rf,open('rf.pkl','wb'))
```
